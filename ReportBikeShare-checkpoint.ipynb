```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
path = 'data\\'
df = pd.read_csv(path + '202112-divvy-tripdata.csv')
```

```python
df.dtypes
```

## Create a column called “ride_length.”

```python
df['ride_length'] = pd.to_datetime(df['ended_at']) - pd.to_datetime(df['started_at'])
```

```python
df
```

## Create a column called “day_of_week,”

```python
# Monday  = 0 , Sunday = 6
df['day_of_week'] = pd.to_datetime(df['started_at']).dt.dayofweek + 2
```

```python
df.describe()
```

## Calculate the mean, max of ride_length

```python
mean_ride_length = df['ride_length'].mean()
mean_ride_length
```

```python
max_ride_length = df['ride_length'].max()
max_ride_length
```

## Calculate the mode of day of week

```python
# Monday  = 0 , Sunday = 6
mode_day_of_week = df['day_of_week'].mode() 
mode_day_of_week
```

```python
df['day_of_week'].value_counts()
```

## Calculate the average ride_length for members and casual riders

```python
df.groupby('member_casual')['ride_length'].agg(lambda x: x.mean())
```

## Calculate the average ride_length for users by day_of_week

```python
df.groupby(['day_of_week', 'member_casual'])['ride_length'].agg(lambda x: x.mean())
```

## Calculate the number of rides for users by day_of_week

```python
df.groupby(['day_of_week', 'member_casual'])['ride_id'].count()
```
